### Key Terms in Pyplot

- **Data Points**: Individual values plotted on a graph, represented as (x, y) coordinate pairs.

- **X-axis**: The horizontal axis of the plot that typically represents the independent variable.

- **Y-axis**: The vertical axis of the plot that typically represents the dependent variable.

- **Figure**: The overall window or page that contains all plot elements. It acts as a container for one or more axes.

- **Axes**: The area within a figure where data is plotted. It includes x-axis and y-axis, ticks, labels, and the plotted data.

- **Plot**: The actual graphical representation of data points on the axes using lines, markers, or other styles.

- **Marker**: A symbol used to represent individual data points on the plot (e.g., circles, squares, triangles).

- **Linestyle**: The style of the line connecting data points (e.g., solid, dashed, dotted).

- **Color**: Specifies the color of plot elements like lines, markers, and text.

- **Legend**: A box or label that identifies different plotted data series by their labels.

- **Label**: A description assigned to a data series, used in the legend and axes.

- **Title**: The heading displayed at the top of the plot, describing the overall content or purpose of the plot.

- **Grid**: A set of intersecting horizontal and vertical lines added to the plot to improve readability and alignment of data points.

- **Functions**: Predefined blocks of code that perform specific tasks. In Pyplot, functions like `plot()`, `xlabel()`, and `show()` are used to create and customize plots.

- **Method**: A function that is associated with an object. In object-oriented Pyplot (e.g., using `Axes`), methods like `ax.plot()` or `ax.set_title()` are used.

- **Parameters**: Inputs passed to functions or methods that control their behavior. For example, `plot(x, y, color='red')` uses `x`, `y`, and `color` as parameters.

- **Keyword Arguments**: Special parameters that are identified by a name (`key=value`) and allow for optional customization in function or method calls.

- **Object-Oriented API**: A more flexible and powerful way to use Matplotlib by working with objects like `Figure` and `Axes`, rather than relying only on state-based functions.

- **DPI (Dots Per Inch)**: Controls the resolution of the figure. Higher DPI means better quality when saving or displaying the plot.

- **Backend**: The rendering engine that Matplotlib uses to draw plots. It can be interactive (e.g., TkAgg, Qt5Agg) or non-inteactive (e.g., Agg for saving to files).

### Common Pyplot Functions and Their Descriptions

- **`plt.grid()`**: Adds a grid to the plot, making it easier to read values by showing horizontal and vertical lines aligned with the ticks.

- **`plt.show()`**: Displays the figure window with all the plots. It is necessary to call this in scripts or programs to render the plot visually.

- **`plt.xlim()`**: Sets or gets the limit of the x-axis. For example, `plt.xlim(0, 10)` will restrict the x-axis from 0 to 10.

- **`plt.ylim()`**: Sets or gets the limit of the y-axis. For example, `plt.ylim(0, 100)` will restrict the y-axis from 0 to 100.

- **`plt.xticks()`**: Sets or gets the tick marks along the x-axis. It can also be used to customize tick labels.

- **`plt.yticks()`**: Sets or gets the tick marks along the y-axis, similar to `xticks()` but for the y-axis.

- **`plt.legend()`**: Displays a legend on the plot, which helps identify different data series when labels are provided.

- **`plt.title()`**: Adds a title to the plot, typically placed at the top of the figure.

- **`plt.plot()`**: The main function to plot data. It draws lines, points, or both, depending on additional arguments like `marker`, `linestyle`, and `color`.

- **`plt.xlabel()`**: Adds a label to the x-axis describing what the axis represents.

- **`plt.ylabel()`**: Adds a label to the y-axis describing what the axis represents.

```python
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales  = [1000,1500,1200,1800]

plt.plot(months,sales, color = 'blue', linestyle = '--', linewidth = 2, marker = '^',
         markerfacecolor = 'red',markeredgecolor = 'red', label = '2025 sales data' )
plt.xlabel('Month')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Report')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.xlim(0,5)
plt.ylim(1000,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

# SUBPLOTS

### Essential Functions in Matplotlib

- **`sharex` / `sharey`**: Parameters used in `plt.subplots()` to allow sharing of the x-axis or y-axis among multiple subplots. This helps in synchronizing the axes across subplots for consistent scaling.

- **`ax.set_title()`**: Sets the title of an individual subplot (Axes object). Useful when working with multiple subplots.

- **`fig.suptitle()`**: Sets a common title for the entire figure, usually placed at the top center. Ideal when the figure contains multiple subplots.

- **`plt.subplots()`**: Creates a figure and a set of subplots (Axes objects). It returns a tuple: `(fig, ax)` or `(fig, axs)` when creating multiple plots.

- **`plt.tight_layout()`**: Automatically adjusts subplot parameters to prevent overlapping of elements like titles, labels, or ticks.

- **`ax.text()`**: Adds custom text at a specified location within an Axes object. Useful for annotations, labels, or explanations directly on the plot.

- **`ax.annotate()`**: Adds annotations with optional arrows pointing to a specific data point. It supports detailed customization of position, style, and arrows.

- **`fig.subplots_adjust()`**: Manually adjusts spacing between subplots in a figure (margins, padding, etc.) using parameters like `left`, `right`, `top`, `bottom`, `wspace`, and `hspace`.

- **`plt.subplot()`**: Used to create a single subplot within a grid layout (e.g., `plt.subplot(2, 2, 1)` creates the first subplot in a 2x2 grid). Unlike `plt.subplots()`, it returns only one Axes at a time.

**GENERAL SYNTAX :**<br>
plt.subplots(nrow, ncols, index)<br>
fig, ax = plt.subplots(nrows, ncols, figsize=(width,height))

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar chart')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y, color = 'r')
ax[0].set_title('Line Plot')

ax[1].bar(x,y, color = 'g')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of line and bar charts')

plt.tight_layout()
plt.show()
```

#### How do you create a scatter plot in Matplotlib? Write a Python script to visualize the relationship between advertising budget and sales revenue of a company. Use: X-axis: Advertising Budget ($),Y-axis: Sales Revenue ($).Use blue circular markers ('bo').Include a regression line if applicable.

```python
x = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y = np.array([25, 45, 50, 65, 70, 85, 95, 110])

plt.scatter(x, y, color='blue', marker='o', label='Data Points')

m, c = np.polyfit(x,y,1)
regression_line = m*x + c
plt.plot(x, regression_line, color='red', label='Regression Line')

plt.xlabel('Advertising Budget')
plt.ylabel('Sales Revenue')
plt.title('Advertising Budget vs Sales Revenue')
plt.legend()

plt.grid()
plt.show()
```

#### What is the difference between a bar chart and a histogram? Explain with examples.

| Feature            | Bar Chart                                          | Histogram                                                |
|--------------------|----------------------------------------------------|-----------------------------------------------------------|
| Purpose            | To compare values across distinct categories       | To show distribution of a continuous variable             |
| Data Type          | Categorical                                        | Numerical (continuous or interval)                        |
| Bars               | Separated by space                                 | Touching each other to indicate continuity                |
| X-Axis             | Categories                                         | Intervals (bins) of continuous data                       |
| Emphasis           | Comparison between categories                      | Frequency distribution / data spread                      |
| Example            | Sales by product type, population by country       | Age distribution, temperature ranges                      |


```python
import matplotlib.pyplot as plt

# Data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
quantities = [10, 15, 7, 12]
scores = [12, 15, 20, 22, 23, 25, 26, 29, 30, 32, 35, 37, 38, 39, 40]

# Create adjacent subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar Chart
axes[0].bar(fruits, quantities, color='skyblue')
axes[0].set_title('Fruit Popularity')
axes[0].set_xlabel('Fruit')
axes[0].set_ylabel('Quantity')
axes[0].grid(True, axis='y', linestyle='--', alpha=0.7)

# Histogram
axes[1].hist(scores, bins=5, color='salmon', edgecolor='black')
axes[1].set_title('Score Distribution')
axes[1].set_xlabel('Score Range')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#### Write a Python script using Matplotlib to create a main plot and an insert the plot with the following constraints:

```python
x = np.linspace(0, 20, 20)
y1 = np.sqrt(x)
y2 = x**3
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(x, y1, color="green", label="Square Root")
ax1.set_xlabel("Values")
ax1.set_ylabel("Square Root")
ax1.set_title("Square Root function")
ax1.legend()
ax2 = fig.add_axes([0.5, 0.2, 0.3, 0.2]) #left,bottom,width,height
ax2.plot(x, y2, color="blue", label="Cube")
ax2.set_xlabel("Values")
ax2.set_ylabel("Cube")
ax2.set_title("Cube function")
ax2.legend()
plt.show()
```

#### Draw aline chart with y = np.sin(x).Fill the pink color in waves

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine Wave', color='blue')
plt.fill_between(x, y, color='red', alpha=0.2)
plt.title("Line Chart with Shaded Area")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid()
plt.show()
```
