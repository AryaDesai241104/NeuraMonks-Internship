## MACHINE LEARNING DAY-11 : K-Nearest Neighbour

### What is KNN ?
**K-Nearest Neighbors (KNN)** is a **supervised machine learning algorithm** used for **classification** and sometimes **regression**. It’s simple, intuitive, and based entirely on the idea of **"similarity"** — things that are similar are likely to have similar outcomes.

---

#### Core Idea

KNN says:

> To classify a new data point, look at the **K closest points** (neighbors) in the training data, and assign the **most common class** among them.

---

#### How It Works (Step-by-Step)

1. **Choose a value for K** (e.g., 3, 5, 7)
2. **Calculate the distance** between the new point and all training points
   (commonly using **Euclidean distance**)
3. **Select the K nearest neighbors** (points with the smallest distances)
4. **Vote**:

   * For **classification**: pick the **most frequent class** among the K neighbors
   * For **regression**: take the **average value** of the K neighbors

---

#### Example (K = 3)

You want to classify a fruit based on its weight and color:

| Weight | Color Score | Fruit  |
| ------ | ----------- | ------ |
| 150g   | 8.5         | Apple  |
| 170g   | 8.9         | Apple  |
| 180g   | 9.2         | Apple  |
| 130g   | 6.5         | Orange |
| 140g   | 6.9         | Orange |

You input a new fruit: **160g, 8.7**

* KNN finds the 3 nearest fruits (maybe 2 Apples and 1 Orange)
* The majority is **Apple**, so it classifies the new fruit as **Apple**

---

#### Common Distance Metrics in KNN

---

##### **Euclidean Distance** (L2 norm)

**Use when**: your features are **continuous numeric values** and the distance is measured "as the crow flies" (i.e., straight-line distance).

##### **Formula**:

$$
D(x, y) = \sqrt{ \sum_{i=1}^{n} (x_i - y_i)^2 }
$$

##### **Example**:

* x = \[1, 2, 3]
* y = \[4, 6, 3]

$$
D = \sqrt{(1-4)^2 + (2-6)^2 + (3-3)^2} = \sqrt{9 + 16 + 0} = \sqrt{25} = 5
$$

---

#### **Manhattan Distance** (L1 norm or Taxicab Distance)

**Use when**: features are numeric, but you care about **axis-aligned** movement (e.g., grid-like data, sparse vectors).

##### **Formula**:

$$
D(x, y) = \sum_{i=1}^{n} |x_i - y_i|
$$

##### **Example**:

* x = \[1, 2, 3]
* y = \[4, 6, 3]

$$
D = |1-4| + |2-6| + |3-3| = 3 + 4 + 0 = 7
$$

---

#### **Minkowski Distance** (Generalized distance)

**Use when**: you want to **tune the distance metric** — it becomes Manhattan when p=1 and Euclidean when p=2.

##### **Formula**:

$$
D(x, y) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{\frac{1}{p}}
$$

##### **Examples**:

* **p = 1** → Manhattan
* **p = 2** → Euclidean
* You can try p = 3 or fractional values for custom behavior.

---

#### **Hamming Distance** (For categorical/binary features)

**Use when**: your data is **categorical** or **binary** and you care about **bitwise or value-wise differences**.

##### **Formula**:

$$
D(x, y) = \sum_{i=1}^{n} [x_i \ne y_i]
$$

(Each mismatched element adds 1)

##### **Example**:

* x = \[1, 0, 1, 1]
* y = \[1, 1, 0, 1]

$$
D = 0 + 1 + 1 + 0 = 2
$$

---

### Summary Table

| Distance      | Formula                     | Best for                    | Type              |                                  |         |
| ------------- | --------------------------- | --------------------------- | ----------------- | -------------------------------- | ------- |
| **Euclidean** | $\sqrt{\sum (x_i - y_i)^2}$ | Continuous numeric features | L2 norm           |                                  |         |
| **Manhattan** | (\sum                       | x\_i - y\_i                 | )                 | Grid-based movement, sparse data | L1 norm |
| **Minkowski** | (\left( \sum                | x\_i - y\_i                 | ^p \right)^{1/p}) | Tunable/general case             | p-norm  |
| **Hamming**   | $\sum [x_i \ne y_i]$        | Binary/categorical data     | Discrete          |                                  |         |

#### When to Use Which:

| Distance Type | Best For                          | Metric Type        |
| ------------- | --------------------------------- | ------------------ |
| Manhattan     | Sparse features, grid layouts     | L1                 |
| Euclidean     | Continuous numeric data           | L2 (Minkowski p=2) |
| Minkowski     | General purpose, tunable          | Generalized        |
| Hamming       | Binary or categorical comparisons | Discrete           |

---

#### Pros

* Very simple and intuitive
* No training phase (lazy learner)
* Works well for small datasets

#### Cons

* **Slow for large datasets** (since it compares with every training point)
* **Sensitive to irrelevant features** and **feature scaling**
* Performance depends heavily on the choice of **K** and **distance metric**

### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the Dataset

```python
df = pd.read_csv('../datasets/Social_Network_Ads.csv')
print(df.sample(10))
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(X)
print('\n')
print(y)
```

### Splitting the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print(X_train)
print('\n')
print(X_test)
print('\n')
print(y_train)
print('\n')
print(y_test)
print('\n')
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print('\n')
print(X_test)
```

### Training the KNN model on Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train,y_train)
```

### Predicting a new Result

```python
y_pred_new = classifier.predict(sc.transform([[30, 87000]]))
print(y_pred_new)
```

### Predicting the test set result

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_pred), 1)), 1))
```

### Making the confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### Visualising the training set results

```python
from matplotlib.colors import ListedColormap

# Inverse scaling to plot in original scale
X_set, y_set = sc.inverse_transform(X_train), y_train

# Choose a reasonable step size
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=1.0),
    np.arange(start=X_set[:, 1].min() - 100, stop=X_set[:, 1].max() + 100, step=500.0)
)

# Predict the classifier's output for every point on the grid
grid_points = np.array([X1.ravel(), X2.ravel()]).T
grid_scaled = sc.transform(grid_points)
Z = classifier.predict(grid_scaled).reshape(X1.shape)

# Plotting decision boundary
plt.contourf(X1, X2, Z, alpha=0.75, cmap=ListedColormap(['red', 'green']))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plotting training points
colors = ['red', 'green']
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], color=colors[i], label=j)

plt.title('KNN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

### Visualising the test set results

```python
# Inverse transform the test features
X_set, y_set = sc.inverse_transform(X_test), y_test

# Choose optimized step sizes
step_x1 = 1.0
step_x2 = 500.0

# Create a grid over the feature space
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=step_x1),
    np.arange(start=X_set[:, 1].min() - 100, stop=X_set[:, 1].max() + 100, step=step_x2)
)

# Predict over the grid and reshape
grid_points = np.array([X1.ravel(), X2.ravel()]).T
Z = classifier.predict(sc.transform(grid_points)).reshape(X1.shape)

# Plot decision boundary
plt.contourf(X1, X2, Z, alpha=0.75, cmap=ListedColormap(['red', 'green']))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Scatter plot for actual test data
colors = ['red', 'green']
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], color=colors[i], label=j)

# Labels and legend
plt.title('KNN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
